```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("parkinsons_updrs.data")
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('motor_UPDRS', axis=1),  # Feature matrix (excluding target)
                                                    data['motor_UPDRS'],  # Target vector
                                                    test_size=0.2,  # 20% for testing
                                                    random_state=42)  # Set random state for reproducibility
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(y_test)
print(y_train)
print(y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
print(y_test)
print(y_train)
print(y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

```

```python
print(y_test)
print(y_train)
print(y_pred)
```

```python
import matplotlib.pyplot as plt


plt.scatter(y_test, y_pred, c='blue', label='Test Values')  
plt.scatter(y_test, y_pred, c='red', alpha=0.7, label='Predicted Values', marker='x') 

plt.xlabel('Actual Target Values')
plt.ylabel('Predicted Target Values')
plt.title('Predicted vs. Actual Values')

plt.legend()

plt.show()

```
